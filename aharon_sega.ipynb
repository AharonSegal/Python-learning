```python
BigList = [["a", "b", "c", "d", "e", "f", "g", "h", "i", 10,2],
           ["one", "two", "three", "four", "five", "six", "seven", "eight","nine","ten","eleven",1,2],
           [1, 2, 3, 4, 5, 6, 7, 8]]


# filter non int from lists (using list comprehension)
def filter_non_int(list_of_sublists):
    return [[item for item in sublist if isinstance(item, int)] for sublist in list_of_sublists]


#1: find largest num in sublists (using list comprehension)
def find_largest_num(list_of_sublists):
    only_int = filter_non_int(list_of_sublists)
    biggest_int = max([num for sublist in only_int for num in sublist])
    print(f"1: The biggest num in all lists is {biggest_int}")

find_largest_num(BigList)

#2: list with largest sum (using max function with sum function as key)
def find_largest_sum(list_of_sublists):
    only_int = filter_non_int(list_of_sublists)
    largest_sum = max(only_int, key=sum)
    print(f"2: {largest_sum} has the largest sum")

find_largest_sum(BigList)

#3: find longest list (using max function with len function as key)
def find_largest_length(list_of_sublists):
    largest_length = max(list_of_sublists, key=len)
    print(f"3: {largest_length} is the longest list")

find_largest_length(BigList)

#4: remove smallest num
def remove_smallest(list_of_sublists):
    only_int = filter_non_int(list_of_sublists)
    # learn what needs to be removed
    to_remove = [min(sublist) for sublist in only_int]

    # remove smallest from lists (enumerate to get value and location)
    for index, list in enumerate(list_of_sublists):
        for item in list:
            if item == to_remove[index]:
                list.remove(item)
    print("4: ", list_of_sublists)

remove_smallest(BigList)

#5: merge content (using list comprehention)
def merge_content(list_of_sublists):
    merged_list = [item for sublist in list_of_sublists for item in sublist]
    print("5: merged list: ", merged_list)

merge_content(BigList)

```

```python
import math

num_my_list = [1,2,3,3,4,1,5,6,7,8,9,20,1,1,1,1,50,20]

#1: find smallest not in my_list [desired = 2]
result_1 = min(num_my_list)+1
print("1: the smallest not in the my_list is" , result_1)

#2: find smallest with one occurrence [desired =4]
def find_smallest_is_once(my_list):
    while True:
        min_num = min(my_list)
        count = my_list.count(min_num)
        if count == 1:
            print(f"2: The smallest number with one occurrence is: {min_num}")
            break
        else:
            while min_num in my_list:
                my_list.remove(min_num)


find_smallest_is_once(num_my_list)

#3: find larges gap between two neighboring nums
####WHY IS THIS WORKING SCINCE 20 - 50 = -30 AND IT IS STILL GIVING 30
def big_gap(my_list):
    gap = 0
    for i in range(len(my_list) - 1):
        x = my_list[i+1] - my_list[i]
        if x > gap:
            gap = x
    print("3: The largest gap between two neighboring numbers is:", gap)
big_gap(num_my_list)

# 4 : longest run
def longest_run(lst):
    max_run = 0
    current_run = 0
    max_run_start_index = 0
    max_run_end_index = 0
    current_run_start_index = 0

    for index in range(len(lst) - 1):
        if lst[index] == lst[index + 1] - 1:
            current_run += 1
            if current_run == 1:
                current_run_start_index = index
            current_run_end_index = index + 1
        else:
            if current_run > max_run:
                max_run = current_run
                max_run_start_index = current_run_start_index
                max_run_end_index = current_run_end_index
            current_run = 0


    if current_run > max_run:
        max_run = current_run
        max_run_start_index = current_run_start_index
        max_run_end_index = current_run_end_index

    longest_sequence = lst[max_run_start_index:max_run_end_index + 1]
    print("4: The longest run in the list is:", longest_sequence)
    print("   Start index:", max_run_start_index)
    print("   End index:", max_run_end_index)


longest_run(num_my_list)

#5: cal average
def cal_average(my_list):
    total_sum = sum(my_list)
    list_length = len(my_list)
    average = total_sum / list_length
    return average

average = cal_average(num_my_list)

# deviation = the average of the variance wich gives the average of the dispertion in the data
def calculate_variance_and_std_deviation(my_list):
    # Calculate the mean "average"
    mean = sum(my_list) / len(my_list)

    # Calculate the squared differences from the mean
    squared_diff = [(x - mean) ** 2 for x in my_list]

    # Calculate the variance
    variance = sum(squared_diff) / len(my_list)

    # Calculate the standard deviation (which is the square root of the variance)
    std_deviation = math.sqrt(variance)

    return variance, std_deviation

variance, std_deviation = calculate_variance_and_std_deviation(num_my_list)

```

```python
import string

my_string = "aaaabbbbbccccccdddddddeffggg"

#1 times for letter
def make_dict_times(my_str):
    letter_dict = {}
    for letter in my_str:
        if letter not in letter_dict:
            letter_dict[letter]= 1
        else:
            letter_dict[letter] += 1
    return(letter_dict)

#2: add 0 for leters that didnt apear at all
def add_non_apear(my_srting):
    the_dict = make_dict_times(my_srting)

    alphabet = string.ascii_lowercase
    for letter in alphabet:
        if letter not in the_dict:
            the_dict[letter] = 0

    print(the_dict)

add_non_apear(my_string)

#3 count letters in sentence
sentence_str = "hello ,it is good to say hello to you , so hello to you"
words_dict = make_dict_times(sentence_str)
print(words_dict)

# BONUS make count for words
def make_dict_word_counts(sentence_str):
    word_dict = {}
    words = sentence_str.replace(",", "").split()

    for word in words:
        if word not in word_dict:
            word_dict[word] = 1
        else:
            word_dict[word] += 1

    print(word_dict)
make_dict_word_counts(sentence_str)
```
